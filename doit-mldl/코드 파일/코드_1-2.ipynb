📌[실습 01-39] 데이터 불러오기

📌[실습 01-40] 데이터 프레임의 특정 열 삭제하기

# 실습 2: tips 데이터

📌[실습 01-  59] 데이터 불러오기

```python
import seaborn as sns
tips = sns.load_dataset("tips")
tips
```

**데이터 전처리**

📌[실습 01-  60] 스케일 표준화하기

```python
# 라이브러리 설치
import seaborn as sns; import matplotlib.pyplot as plt;
from sklearn.preprocessing import StandardScaler

# StandardScaler 적용
scaler_SS = StandardScaler()
tips_ss = tips.copy()
tips_ss["total_bill"] = scaler_SS.fit_transform(tips[["total_bill"]])

# 확인
print("원본 데이터 통계")
print(tips["total_bill"].describe())

print("\nStandardScaler 적용 후")
print(tips_ss["total_bill"].describe())
```

📌[실습 01-  61] 스케일 표준화 차이 비교하기

```python
plt.figure(figsize=(10, 3))     # 그래프 전체 크기 설정

# 원본 total_bill
plt.subplot(1,2,1)   # 그림 프레임 위치 설정
sns.histplot(tips["total_bill"], kde=True, color="skyblue")
plt.title("원본 total_bill")
plt.xlabel("총 지불액")

# Standard 전처리 후 total_bill
plt.subplot(1,2,2)
sns.histplot(tips_ss["total_bill"], kde=True, color="salmon")
plt.title("StandardScaler 적용 후 total_bill")
plt.xlabel("표준화된 총 지불액")
plt.show()
```

📌[실습 01-  62] 스케일 정규화 전후 비교하기

```python
from sklearn.preprocessing import MinMaxScaler
scaler_mm=MinMaxScaler() # MinMaxScaler 적용객체 생성
tips_mm=tips.copy()
tips_mm["total_bill"]=scaler_mm.fit_transform(tips[["total_bill"]]) # 정규화 실행

print("원본 데이터 통계")
print(tips["total_bill"].describe())
print("\nMinMaxScaler 적용 후")
print(tips_mm["total_bill"].describe())

plt.figure(figsize=(6, 10))     # 그래프 전체 크기 설정
# 원본 total_bill
plt.subplot(2,1,1)   # 그림 프레임 위치 설정
sns.histplot(tips["total_bill"], kde=True, color="skyblue")
plt.title("원본 total_bill")
plt.xlabel("총 지불액")

# MinMax 전처리 후 total_bill
plt.subplot(2,1,2)
sns.histplot(tips_mm["total_bill"], kde=True, color="salmon")
plt.title("MinMaxScaler 적용 후 total_bill")
plt.xlabel("정규화된 총 지불액 (0~1)")
plt.show()
```

# 사용자 정의 함수 만들기

📌[실습 01-  63] 새로운 함수 만들기

```python
def min_max(data):         # (1) 함수 정의와 입력값 설정
    q=(data - data.min()) / (data.max() - data.min())  # (2) 연산 규칙
    return q               # (3) 결과 반환
```

📌[실습 01-  64] 함수 호출하기

```python
z=min_max(tips["total_bill"]) # 함수 호출

print("min_max 함수로 정규화된 값"); print(z); print()
print("정규화된 값의 기초 통계")
print(z.describe())
```
