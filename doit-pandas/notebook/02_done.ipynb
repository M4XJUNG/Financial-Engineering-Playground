## loc 속성으로 행 단위 데이터 추출하기(35쪽)

image.png

```python
# 인덱스
'''
실행했을 때 맨 앞에 0, 1, 2, 3, 4 가 인덱스를 의미.
행 번호는 데이터의 순서를 따라가기 때문에 정수만으로 데이터를 조회하거나
추출할 수 있으며 실제 데이터프레임에서는 확인할 수 없는 값이다.
'''
df.head()
```

```python
print(df.loc[0])
```

```python
print(df.loc[99])
```

```python
# 인덱스에 없는 값을 사용하면 오류가 발생한다.
print(df.loc[-1])
```

```python
# 데이터프레임의 마지막 행 데이터를 추출하려면
# 마지막 행 데이터의 인덱스를 알아야함.
number_of_rows = df.shape[0] # 행 크기
# 인덱스가 0부터 세므로 - 1을 해줘야 마지막 행을 의미.
last_row_index = number_of_rows - 1
print(df.loc[last_row_index])
```

```python
# tail 메서드에 인자 n에 1을 전달하면 마지막 행의 데이터를 추출 가능.
print(df.tail(n=1))
```

```python
# 2를 전달하면 2개의 데이터를 추출할 수 있다.
print(df.tail(n=2))
```

```python
# 인덱스가 0, 99, 999인 데이터를 한 번에 추출하려면
# 리스트에 원하는 인덱스를 담아 loc 속성에 전달
print(df.loc[[0, 99, 999]])
```

## tail 메서드와 loc 속성이 반환하는 자료형은 서로 달라요!(37쪽)

```python
# tail 메서드와 loc 속성이 반환하는 자료형은 서로 다르다.
# loc 속성이 반환한 데이터 자료형은 시리즈.
# tail 메서드가 반환한 데이터 자료형은 데이터프레임.
subset_loc = df.loc[0]
subset_tail = df.tail(n=1)

print(type(subset_loc))
print(type(subset_tail))
```

## 그룹화한 데이터의 평균 구하기(44쪽)
### 1. lifeExp 열을 연도별로 그룹화하여 평균 계산하기

```python
print(df.head(n=10))
```

```python
# lifeExp 열을 연도별로 그룹화하여 평균 계산하기
'''
예를 들어 연도별 lifeExp 열의 평균을 계산하려면?
데이터를 year열로 그룹화하고 lifeExp 열의 평균을 구하면 된다.
데이터프레임의 groupby 메서드에 year 열을 전달하여
연도별로 그룹화한 다음 lifeExp 열을 지정하여 mean 메서드로 평균 구함.
'''
print(df.groupby('year')['lifeExp'].mean())
```

```python
# 위에서 작성한 코드를 작은 단위로 나누어 살펴보자.
grouped_year_df = df.groupby('year')
print(type(grouped_year_df))
```

```python
# 위에서 얻은 데이터프레임이 저장된 메모리의 위치를 알 수 있음.
print(grouped_year_df)
```

```python
# 이어서 lifeExp 열을 추출한 결과를 살펴보자.
grouped_year_df_lifeExp = grouped_year_df['lifeExp']
print(type(grouped_year_df_lifeExp))
```

```python
# 마지막으로 평균을 구하는 mean 메서드를 사용한 결과를 살펴보자.
mean_lifeExp_by_year = grouped_year_df_lifeExp.mean()
print(mean_lifeExp_by_year)
```

### 6. lifeExp, gdpPercap 열의 평균값을 연도, 지역별로 그룹화하여 한 번에 계산하기

```python
multi_group_var = df.groupby(['year', 'continent'])[['lifeExp', 'gdpPercap']].mean()
print(multi_group_var)
```

```python
print(type(multi_group_var))
```

### 7. 그룹화한 데이터의 개수 세기

```python
# 데이터의 빈도수는 nunique 메서드를 사용하면 쉽게 구할 수 있다.
print(df.groupby('continent')['country'].nunique())
```
